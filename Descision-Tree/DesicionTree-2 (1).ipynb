```
#Imporing the module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Loading the dataset
df=pd.read_csv('tic-tac-toe-endgame.csv')
```

```
#Printing the first 10 rows of the dataset
df.head(10)
```

```
#Giving labels to each values in the column
from sklearn.preprocessing import LabelEncoder
#Creating objects for the class
l_v1=LabelEncoder()
l_v2=LabelEncoder()
l_v3=LabelEncoder()
l_v4=LabelEncoder()
l_v5=LabelEncoder()
l_v6=LabelEncoder()
l_v7=LabelEncoder()
l_v8=LabelEncoder()
l_v9=LabelEncoder()
l_v10=LabelEncoder()
```

```
#Creating new columns based on labels for each values
df['V1_n']=l_v1.fit_transform(df['V1'])
df['V2_n']=l_v2.fit_transform(df['V2'])
df['V3_n']=l_v3.fit_transform(df['V3'])
df['V4_n']=l_v4.fit_transform(df['V4'])
df['V5_n']=l_v5.fit_transform(df['V5'])
df['V6_n']=l_v6.fit_transform(df['V6'])
df['V7_n']=l_v7.fit_transform(df['V7'])
df['V8_n']=l_v8.fit_transform(df['V8'])
df['V9_n']=l_v9.fit_transform(df['V9'])
df['V10_n']=l_v10.fit_transform(df['V10'])
```

```
#Loading the dataset
df
```

```
#Taking input and output columns
inp=df[['V1_n','V2_n','V3_n','V4_n','V5_n','V6_n','V7_n','V8_n','V9_n']]
out=df[['V10_n']]
inp
```

```
#Creating training and testing dataset
from sklearn.model_selection import train_test_split as tts
train_input,test_input,train_output,test_ouptut=tts(inp,out,train_size=0.9,random_state=10)
train_input
```

```
#Solving using desicion tree classisifier
from sklearn import tree
des=tree.DecisionTreeClassifier()
des.fit(train_input,train_output)
```

```
#Predicting output with random values(x,x,x,0,0,0,b,b,b)
des.predict([[2,2,2,1,1,1,0,0,0,]])
```

```
#Prediciting the ouput with testing dataset
des.predict(test_input)
```

```
#Actual output of testing dataset
test_ouptut
```

```
#Comparing the efficiency between actual and predicted output of test dataset
des.score(test_input,test_ouptut)
#92% efficiecny
```

```
##Comparing the efficiency between actual and predicted output of training dataset
des.score(train_input,train_output)
#100% Efficiency
```
